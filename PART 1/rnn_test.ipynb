```python
import numpy as np
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical
```

```python

# Generate some example data
text = "hello world hello world hello world hello world"
chars = list(set(text))
char_to_index = {ch: i for i, ch in enumerate(chars)}
index_to_char = {i: ch for ch, i in char_to_index.items()}

# Prepare the dataset
SEQUENCE_LENGTH = 10
X_data, y_data = [], []

for i in range(0, len(text) - SEQUENCE_LENGTH):
    sequence = text[i:i + SEQUENCE_LENGTH]
    label = text[i + SEQUENCE_LENGTH]
    X_data.append([char_to_index[char] for char in sequence])
    y_data.append(char_to_index[label])

X = np.reshape(X_data, (len(X_data), SEQUENCE_LENGTH, 1))
X = X / float(len(chars))
y = to_categorical(y_data)

# Build the RNN model
model = Sequential()
model.add(SimpleRNN(32, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(SimpleRNN(32))
model.add(Dense(y.shape[1], activation='softmax'))

# Compile and train the model
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=100, batch_size=16)

# Test the model with a seed sequence
seed = "hello worl"
seed_encoded = np.array([[char_to_index[char] for char in seed]])
seed_encoded = np.reshape(seed_encoded, (1, len(seed), 1))
seed_encoded = seed_encoded / float(len(chars))

# Predict the next character
predicted_output = model.predict(seed_encoded, verbose=0)
predicted_index = np.argmax(predicted_output[0])
predicted_char = index_to_char[predicted_index]

print(f"Seed: {seed}")
print(f"Predicted next character: {predicted_char}")
```
